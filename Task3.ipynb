```python
# Removing outliers

column = df[['age','campaign','duration']]
q1 = np.percentile(column, 25)
q3 = np.percentile(column, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
df[['age','campaign','duration']] = column[(column > lower_bound) & (column < upper_bound)]
```

```python
high_corr_cols = ['emp.var.rate','euribor3m','nr.employed']
```

```python
# copy with original dataframe
df1 = df.copy()
df1.columns
```

```python
# Removing high correlated columns from the dataset
df1.drop(high_corr_cols,inplace=True,axis=1)  # axis=1 indicates columns
df1.columns
```

```python
# showing dimensions of the updated dataset
df1.shape
```

```python
# Conversion of categorical columns into numerical columns using label encoder.
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df_encoded = df1.apply(lb.fit_transform)
df_encoded
```

```python
df_encoded['deposit'].value_counts()
```

```python
x = df_encoded.drop('deposit',axis=1)  # independent variable
y = df_encoded['deposit']              # dependent variable
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
print(4119*0.25)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
def eval_model(y_test,y_pred):
    acc = accuracy_score(y_test,y_pred)
    print('Accuracy_Score',acc)
    cm = confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))

def mscore(model):
    train_score = model.score(x_train,y_train)
    test_score = model.score(x_test,y_test)
    print('Training Score',train_score)
    print('Testing Score',test_score)
```

```python
# Importing  Decision Tree library
from sklearn.tree import DecisionTreeClassifier
```

```python
# Building Decision Tree Classifier Model
dt = DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
# Evaluating training and testing accuracy
mscore(dt)
```

```python
# Generating prediction
ypred_dt = dt.predict(x_test)
print(ypred_dt)
```

```python
# # Evaluate the model - confusion matrix, classification Report, Accuaracy
eval_model(y_test,ypred_dt)
```

```python
#Copy for parsing
bank_data = df.copy()
```

```python
#Exploring People who made a deposit Vs Job category
jobs = ['management','blue-collar','technician','admin.','services','retired','self-employed','student',\
        'unemployed','entrepreneur','housemaid','unknown']

for j in jobs:
    print("{:15} : {:5}". format(j, len(bank_data[(bank_data.deposit == "yes") & (bank_data.job ==j)])))
```

```python
#Different types of job categories and their counts
bank_data.job.value_counts()
```

```python
#Combine similar jobs into categiroes
bank_data['job'] = bank_data['job'].replace(['management', 'admin.'], 'white-collar')
bank_data['job'] = bank_data['job'].replace(['services','housemaid'], 'pink-collar')
bank_data['job'] = bank_data['job'].replace(['retired', 'student', 'unemployed', 'unknown'], 'other')
```

```python
#New value counts
bank_data.job.value_counts()
```

```python
#Displaying the count of unique values in the 'poutcome' column
bank_data.poutcome.value_counts()
```

```python
#Combining 'unknown' and 'other' as 'other' isn't really match with either 'success' or 'failure'
bank_data['poutcome'] = bank_data['poutcome'].replace(['other'] , 'unknown')
bank_data.poutcome.value_counts()
```

```python
#Dropping 'contact', as every participant has been contacted.
bank_data.drop('contact', axis=1, inplace=True)
```
